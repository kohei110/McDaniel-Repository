### 6. Vector Operations and Basic Calculations
   
Given 1D arrays A = [4, 9, 16, 25, 36] and B = [2, 3, 4, 5, 6]:

(b) Create array D by calculating B squared  
(c) Create array E by adding A and B element-wise
