### Q17. Define a function to filter all passengers with ticket Fare less than 20 dollars. Use apply function of pandas to the dataset, display the result, and compute the average fare from the result.

```python
def cheap_fare_filter(fare):
    if fare < 20:
        return True
    else:
        return False
    
cheap_fare = raw_data[raw_data['Fare'].apply(cheap_fare_filter)]

print(cheap_fare['Fare'].mean())
```

### Q19(Y2). Which attribution has higher correlation with Survived attribution?

```python
import seaborn as sns
import matplotlib.pyplot as plt

#raw_data.info()
categorical_attr = ['Sex', 'Embarked','Name', 'Ticket', 'Cabin']
raw_data[categorical_attr].nunique()

# drop meaningless non numeric attribution since those attribution have too many unique values
categorical_attr = ['Sex', 'Embarked']
dummies = pd.get_dummies(raw_data, columns=categorical_attr,dtype=float)
drop_attr = ['Name', 'Ticket', 'Cabin']
numerical_data = dummies.drop(columns=drop_attr)


# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(numerical_data.corr(), annot=True, fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features", fontsize=16)
plt.show()

# Correlation with 'Survived', excluding 'Survived' itself
correlation_with_survived = numerical_data.corr()['Survived'].drop('Survived')

# Seaborn Bar Chart for Correlation with 'Survived'
plt.figure(figsize=(10, 6))
sns.barplot(
    x=correlation_with_survived.sort_values(ascending=False).index,
    y=correlation_with_survived.sort_values(ascending=False).values
)
plt.title("Correlation of Features with Survived", fontsize=16)
plt.ylabel("Correlation Coefficient", fontsize=12)
plt.xlabel("Features", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

#### Answer: Female has the highest correlation with Survived

### Q20(Y3). Looks like Femal has highest correlation against Survived. Did female survival rates vary across passenger classes?



```python
gender_class_survival = raw_data.groupby(['Sex', 'Pclass'])['Survived'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=gender_class_survival, x='Pclass', y='Survived', hue='Sex')

plt.title('Survival Rates by Gender and Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.legend(title='Gender', fontsize=10)
plt.ylim(0, 1) 
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#### Answer: Female survival rate across the three passenger classes is different
