### *Optional*: Q19: Display the age ranks of all passengers in the dataset using the categories where the whole age range of all passengers is divided into equal-length.
<pre>
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
</pre>
<img src="http://mathqs.mcdaniel.edu/ANA522/mod03/img/titanic_age_bin.png" 
     width="600" height="900" alt="attributes list" />

```python
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
bins = [18, 25, 35, 60, 100]
```

```python
min_age = titanic['Age'].min() 
max_age = titanic['Age'].max()

bin_width = (max_age - min_age) / 4
bins = [min_age, min_age + bin_width, min_age + 2*bin_width, min_age + 3*bin_width, max_age]

pd.cut(titanic['Age'], bins=bins, labels=group_names).value_counts()
```
